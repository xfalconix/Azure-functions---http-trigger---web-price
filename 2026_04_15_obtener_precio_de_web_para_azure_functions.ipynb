```python
import requests
import json
import pandas as pd
import datetime
```

```python
url = 'https://www.aftgrupo.com/productos/ref-14062510'

# Realizar la solicitud HTTP
try:
    response = requests.get(url)
    response.raise_for_status() # Lanza una excepción para errores HTTP
    html_content = response.text
    print(f"Página obtenida correctamente: {url}")
except requests.exceptions.RequestException as e:
    print(f"Error al obtener la página: {e}")
    html_content = None
```

```python
product_name = 'No encontrado'
product_price = 'No encontrado'

if html_content:
    # Extraer el nombre del producto
    start_name_tag = '<strong>DESCRIPCIÓN:</strong><br>'
    end_name_tag = '<br />'
    name_start_idx = html_content.find(start_name_tag)
    if name_start_idx != -1:
        name_start_idx += len(start_name_tag)
        name_end_idx = html_content.find(end_name_tag, name_start_idx)
        if name_end_idx != -1:
            product_name = html_content[name_start_idx:name_end_idx].strip()
            product_name = product_name.replace('\r\n', '').strip()

    # Extraer el precio del producto
    start_price_tag = '<div class="precio_ficha">'
    end_price_tag = '</div>'
    price_start_idx = html_content.find(start_price_tag)
    if price_start_idx != -1:
        price_start_idx += len(start_price_tag)
        price_end_idx = html_content.find(end_price_tag, price_start_idx)
        if price_end_idx != -1:
            product_price = html_content[price_start_idx:price_end_idx].strip()

    print(f"Nombre del producto: {product_name}")
    print(f"Precio del producto: {product_price}")
else:
    print("No se pudo extraer información sin el contenido HTML.")
```

```python
# Guardar en una variable el producto el precio y la hora de consulta
data_to_save = {
    'Producto': product_name,
    'Precio': product_price,
    'Timestamp': datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
}
```
